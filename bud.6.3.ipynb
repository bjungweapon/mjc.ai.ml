```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/lynx.csv")
```

```python
print(df.head(10))
```

```python
print(df.describe())
```

```python
#10년 단순 이동 평균.   rolling(n).mean() 
```

```python
df["sma"] = df["value"].rolling(10).mean()
plt.plot(df["value"])
plt.plot(df["sma"])
plt.show()
```

```python
#rolling() : 기간 이동  
#rolling method는 현재 열에 대해여 일정 ㅋ기의 창을 이용하여 windows값을 이동하는것. 
#rollingmax , rollingmin
```

```python
df["ewm"] = df["value"].ewm(10).mean()
plt.plot(df["value"])
plt.plot(df["ewm"])
plt.show()
```
