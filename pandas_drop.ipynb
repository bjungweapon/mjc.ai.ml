```python

df = pd.DataFrame(
    {'name': ['KIM', 'LEE', 'SMITH','BROWN', 'MILLER'],
     'age': [24, 32, 43, 24, np.nan],
     'height': [178, 168, 171, 185, 176],
     'sex': ['M', 'F', 'F', 'M', 'F']})
df
```

```python
df.drop([1,2], axis=0)

## axis = 0 (세로) 축을 기준으로 해서 0 과 1을 지워라.
```

```python
df.drop(['age', 'height'], axis=1)
## 가로축을 기준으로 해서,  'age',와 'height'를 지워라.
```

```python
df
```

```python
df[['age', 'height']].mean(axis=0)

## axis = 0 (세로축)을 기준으로 해서 'age'와 'height'의 평균을을 구하라.
```
